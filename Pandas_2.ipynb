```python
 df = pd.read_csv('weather_data (1).csv', parse_dates=['day'])
```

```python
df.set_index('day',inplace = True)
```

```python
df
```

```python
# filling missing Value 
fill_df = df.fillna(0)
```

```python
fill_df
```

```python
## fill by dict
dict_df = df.fillna({
     'temperature':0,
    'windspeed':0,
    'event':'No Event'
     
})
```

```python
dict_df
```

```python
## Carry Forwarding Previous value 
ffill_df = df.fillna(method = 'ffill')  # we can use also backward fill by "bfill"
```
